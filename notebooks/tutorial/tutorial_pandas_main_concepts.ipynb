# Core Data Structures

**1.) Series:**
a one-dimensional labeled array holding data of any type
such as integers, strings, Python objects etc.

**2.) DataFrame:**
a two-dimensional data structure that holds data like a two-dimension array or a table with rows and columns.



**1.) Series**

https://pandas.pydata.org/docs/user_guide/dsintro.html#series

basic method to create a series:

s = pd.Series(data, index=index)

Here, data can be many different things:
- List, tuple, or array-like
- NumPy ndarray (np.array([1, 2, 3]))

- Dictionary (Keys are used as index, values are used as data)
- Scalar value, constant of certain type (index needs to be provided as parameter)
- Existing pandas series(This creates a new series)

Default behavior for Index:

 RangeIndex(start=0, stop=n, step=1) where n is the length of the Series.

```python
ser1 = pd.Series([1, 3, 5, 6, 8])
ser1
```

By default index is create in range as if it were list index

```python
ser1.index
```

```python
ser1[2]
```

Mixed types get converted into objects

```python
ser2 = pd.Series([1, 'v', '5', 6, 8])
ser2
```

You can also specifiy index of columns

```python
ser3 = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
ser3
```

```python
ser3.index
```

```python
ser3['a'], ser3['b'], ser3['c']
```

**How to reset index of series**

When you have custom index, you can reset index back to default range

```python
series_custom_index = pd.Series([10, 20, 30], index=["a", "b", "c"], name="values")
series_custom_index
```

Defult behavour when using reset_index, is that a new dataframe will be created, with the index as column.

```python
out = series_custom_index.reset_index(drop=False)
out
```

```python
type(out) is pd.DataFrame
```

```python
out['index']
```

If you decide to drop the index a series is returned instead of dataframe.

```python
out = series_custom_index.reset_index(drop=True)
out
```

```python
type(out) is pd.core.frame.Series
```

**2.) Data Frame**

https://pandas.pydata.org/docs/user_guide/dsintro.html#dataframe


 Example of creation of datafram from dictionary of objects, where key is column labels and values are column values**

```python
# DataFrame
df = pd.DataFrame(
    {
        "A": list(range(4)),
        "B": pd.Timestamp("20130102"),
        "C": pd.Series(1, index=list(range(4)), dtype="float32"),
        "D": np.array([3] * 4, dtype="int32"),
        "E": pd.Categorical(["test", "train", "test", "train"]),
        "F": "foo",
    }
)
```

```python
df
```

```python
df.index
```

```python
df.columns
```

Each column has its own separate type

```python
df.dtypes
```

**Accesing columns via single or double brackets**

 	Single brackets  Series

 	Double brackets  DataFrame

Accessed column from dataframe is a Series, when using single brackets.

```python
df['A']
```

```python
type(df['A']) is pd.Series
```

Accessed column from dataframe is a Dataframe, when using double brackets.

```python
df[['A']]
```

```python
type(df[['A']]) is pd.DataFrame
```

**2.) Saving of dataframe to csv file**

You can decide wheather to save to dataframe with index or no

Important parameters:

- sep, is the columns separator.
- index , by default false, exports also index.
- header , by default true, also exports column names as header.
- columns, by default all, limit export to only certain columns.
- mode, "w" is for write, "a" is for append.

```python
# DataFrame
df = pd.DataFrame({
		'Name': ['Alice', 'Bob', 'Charlie'],
		'Age': [25, 30, 35],
		'Score': [85, 90, 95]
})


# Save to file
df.to_csv("data.csv", sep=',', index=False, header=True,  columns=['Name', 'Score'])
```

**You can also create compressed csv file in a zip fil**

```python
compression_opts = dict(method='zip',
                        archive_name='out.csv')
df.to_csv('out.zip', index=False,
          compression=compression_opts)
```

# Indexing into dataframe

- .loc[]: Label-based selection.
- .iloc[]: Position-based selection.
- Boolean indexing: df[df['column'] > value]


```python
	# DataFrame
df = pd.DataFrame({
		'Name': ['Alice', 'Bob', 'Charlie'],
		'Age': [25, 30, 35],
		'Score': [85, 90, 95]
  })
```

```python
df
```

```python
# Select column
df['Age']
```

```python
# Filter rows
df[df['Score'] >= 90]
```

```python
# Select row by label
df.loc[1]
```

```python
# Select row by position
df.iloc[2]
```

```python
type(df['Age']), type(df.iloc[2])
```

```python
type(df[df['Score'] >= 99])
```

```python
type(df[df['Score'] >= 90])
```
